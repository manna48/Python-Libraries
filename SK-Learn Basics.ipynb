### Generating Datasets

- Some useful functions to generate datasets available in sklearn.datasets

- make_regression
- make_classification
- make_blobs
- make_moons

```python
from sklearn.datasets import make_regression,make_classification,make_blobs,make_moons
```

```python
from matplotlib import pyplot as plt
```

```python
X,Y = make_regression(n_samples=400,n_features=1,n_informative=1,noise=10.8)
```

```python
print(X.shape,Y.shape)
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
XC,YC = make_classification(n_samples=400,n_features=2,n_classes=3,n_clusters_per_class=1,n_informative=2,n_redundant=0,random_state=5)
```

```python
print(XC.shape,YC.shape)
```

```python
plt.scatter(XC[:,0],XC[:,1],c=YC)
plt.show()
```

### Blobs Dataset

```python
XB,YB = make_blobs(n_samples=1000, n_features=2, centers=5,random_state=10)
```

```python
plt.scatter(XB[:,0],XB[:,1],c=YB)
plt.show()
```

### Moons Dataset

```python
XM,YM = make_moons(n_samples=400,noise=0.2, random_state=5)
```

```python
plt.scatter(XM[:,0],XM[:,1],c=YM)
plt.show()
```

### Subplots

```python
plt.figure(figsize=(8,8))

plt.subplot(221)
plt.title("Regression")
plt.scatter(X,Y)

plt.subplot(222)
plt.title("Classification")
plt.scatter(XC[:,0],XC[:,1],c=YC)



plt.subplot(223)
plt.title("Blobs")
plt.scatter(XB[:,0],XB[:,1],c=YB)


plt.subplot(224)
plt.title("Moons")
plt.scatter(XM[:,0],XM[:,1],c=YM)

plt.show()
```

```python
mnist = load_digits()
```

```python
X = mnist.data
Y = mnist.target
print(X.shape)
print(Y.shape)
```

```python
from matplotlib import pyplot as plt
plt.imshow(X[3].reshape(8,8),cmap='gray')
plt.show()
```

```python
X,Y = make_classification(n_samples =100,n_features=2,n_informative=2,n_redundant=0,n_classes=2)

plt.figure(0,figsize=(8,8))
print(X.shape)
print(Y.shape)

plt.subplot(221)
plt.scatter(X[:,0],X[:,1],c=Y,edgeColor='k')

plt.subplot(222)
X1, Y1 = make_classification(n_samples=200,n_features=2, n_redundant=0, n_informative=2,
                             n_clusters_per_class=1, n_classes=3)

plt.scatter(X1[:,0],X1[:,1],c=Y1,edgeColor='k')

plt.subplot(223)
X2, Y2 = make_blobs(n_features=2, centers=3)
plt.scatter(X2[:,0],X2[:,1],c=Y2+1,edgeColor='k')

plt.subplot(224)
plt.title("Regression")
X3, Y3= make_regression(n_features=1,n_samples=300,bias=4,noise=8)
print(X3.shape,Y3.shape)
plt.scatter(X3,Y3,color='green')

plt.show()
```
